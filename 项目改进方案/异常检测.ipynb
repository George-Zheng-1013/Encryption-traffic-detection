要实现“Unknown 样本池 -> 聚类 -> 人工标注 -> 增量再训练 -> 新模型上线”的自动流程，本质上是在构建一个**主动学习 (Active Learning) 与 MLOps 数据闭环**。这在创业大赛中是非常亮眼的“商业壁垒”——你的系统不仅能发现未知，还能不断自我进化。

针对你的 CNN 28×28 灰度图分类模型，以下是落地这个闭环的具体架构设计和推荐工具：

### 一、 系统闭环架构设计

整个流程可以拆分为 5 个核心模块，建议采用 Python 脚本定时任务（Cron）或轻量级工作流引擎来串联：

#### 1. Unknown 样本池 (Data Lake)
*   **实现：** 当在线推理引擎（NetGuard）判定某个流的置信度低于阈值，或特征距离过远时，将其标记为 `Unknown`。
*   **存储：** 将该流的原始特征（28×28 图片/Numpy 数组）、抽取的中间层 Embedding 向量（如 CNN 倒数第二层的 128 维向量），以及 5-tuple 信息保存到数据库中（推荐 **MongoDB** 或 **Elasticsearch**，方便检索）。

#### 2. 自动聚类 (Clustering)
*   **实现：** 定期（例如每天凌晨）触发聚类脚本。读取最近一周积累的 Unknown Embedding 向量。
*   **算法推荐：** 使用 **HDBSCAN** 或 **DBSCAN**（通过 `scikit-learn` 或 `hdbscan` 库）。
    *   *为什么不用 K-Means？* 因为你不知道到底有几种新的未知流量，HDBSCAN 不需要预先指定聚类数量 K，且能自动过滤离群点（Noise）。
*   **效果：** 聚类后，你会发现 Unknown 样本被分成了 `Cluster_1`（可能是某种新型 P2P 下载）、`Cluster_2`（可能是新型木马）以及散乱的孤立点。

#### 3. 人工标注接口 (Human-in-the-Loop)
*   **实现：** 不要自己从头写前端！强烈推荐使用开源标注工具 **Label Studio**。
*   **对接：** 编写一个 Python 脚本，将聚类好的大簇（如 Cluster_1 选出 10 张代表性的 28×28 灰度图转成 PNG）推送到 Label Studio 的 API。
*   **人工操作：** 安全运营人员（分析师）在网页上看到这批极度相似的图，结合抓包记录，将其打上新标签（如 `New_Malware_X`）。

#### 4. 增量再训练 (Incremental Retraining)
*   **挑战：** 灾难性遗忘（Catastrophic Forgetting）——只用新数据训练，模型会忘了老数据。
*   **实现：**
    *   **回放机制 (Replay Buffer)：** 维护一个“核心代表集”（Core-set），包含所有旧类别的少量高质量样本。每次训练时，把人工标注好的新数据和回放集混合在一起训练。
    *   **微调 (Fine-tuning)：** 将 CNN 的全连接分类头（Classifier Head）从 $N$ 分类扩展为 $N+1$ 分类，冻结卷积层，先训练分类头，再全量微调。

#### 5. 新模型上线 (Model Deployment)
*   **实现：** 训练脚本完成后，将新模型导出为 ONNX 格式，覆盖特定目录下的旧模型文件，并更新版本号。
*   **热更新：** 在线推理引擎（如 FastAPI 或 Triton Inference Server）监控该目录，发现新版本后，自动加载新模型，实现业务不中断的平滑升级。
