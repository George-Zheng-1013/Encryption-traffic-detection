下面给你一套从 0 到可跑通的 PyTorch 详细流程，按 Windows + PowerShell 写。

**1) 环境配置**
- 进入项目根目录（你当前就是这个目录也可以）  
  cd d:/HP/OneDrive/Desktop/学校/项目/优本项目/文章+代码/costSensitive-new/costSensitive
- 创建并激活虚拟环境  
  py -3.10 -m venv .venv  
  .\.venv\Scripts\Activate.ps1
- 升级基础工具  
  python -m pip install -U pip setuptools wheel
- 安装 PyTorch（CPU 版，最稳）  
  pip install torch torchvision torchaudio
- 安装项目依赖（PyTorch 路线需要）  
  pip install numpy pandas tqdm matplotlib scikit-learn pillow opencv-python scapy dpkt gevent
- 验证安装  
  python -c "import torch; print(torch.__version__); print('cuda=', torch.cuda.is_available())"

**2) 数据集准备（先走最短可运行路径）**
- 你的仓库已包含可直接训练的数据目录 5_Mnist-10，并且脚本默认就读取这里。  
  训练读取位置见 RSG_pytroch_CNN784.py  
  推理读取位置见 inference.py
- 先确认这 4 个文件存在：  
  5_Mnist-10/train-images-idx3-ubyte  
  5_Mnist-10/train-labels-idx1-ubyte  
  5_Mnist-10/t10k-images-idx3-ubyte  
  5_Mnist-10/t10k-labels-idx1-ubyte  
  以及对应 .gz 版本（脚本实际读取 .gz）。

**3) 如果你要“从 pcap 开始”重建数据（可选）**
- 非 VPN 流量：运行 PcaptoImage.py（生成 pcap 切片 + png）  
- VPN 流量：运行 PcaptoImage-vpn-all.py
- 然后运行 mnist.py 把 png 打包成 IDX，并压缩成 .gz 给训练脚本使用。  
- 注意：mnist.py 里用了 gzip 命令，Windows 若无 gzip 命令会失败；这时建议先走“现成 5_Mnist-10 数据”路线。

**4) 训练（PyTorch）**
- 运行训练：  
  python RSG_pytroch_CNN784.py
- 训练结束会保存模型到 pytorch_model/convnet.pth，保存逻辑在 RSG_pytroch_CNN784.py。

**5) 推理与结果导出**
- 运行推理：  
  python inference.py
- 推理会加载 pytorch_model/convnet.pth，并输出 pytorch_model/predictions.csv，对应代码在 inference.py。

**6) 常见问题（你大概率会遇到）**
- 模型文件不存在：先确认训练是否成功生成 pytorch_model/convnet.pth。
- 数据文件不存在或不是 .gz：检查 RSG_pytroch_CNN784.py 读取的是 .gz 名称。
- CUDA 不可用：不影响运行，脚本会自动回退 CPU（见 RSG_pytroch_CNN784.py）。
