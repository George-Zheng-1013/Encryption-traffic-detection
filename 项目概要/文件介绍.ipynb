**主线（你已经跑通的）**
- RSG_pytroch_CNN784.py：PyTorch 训练脚本（读取 `5_Mnist-10`、训练 CNN、保存权重、写 `train_log.csv`）。
- inference.py：推理脚本（加载权重，生成样本级预测 `predictions.csv`）。
- 5_Mnist-10：训练/测试数据（MNIST IDX 及 `.gz`，当前训练/推理直接使用）。
- pytorch_model：主线输出目录，包含模型与结果：
  - convnet.pth：训练后模型权重
  - train_log.csv：epoch 级训练日志
  - predictions.csv：推理逐样本结果

**数据预处理链（从流量到图像/IDX）**
- PcaptoImage.py：非 VPN 流量处理（pcap 切片、定长补齐、转 28×28 图）。
- PcaptoImage-vpn-all.py：VPN 流量处理版本。
- PcaptoImage2.py：另一套类别配置/采样参数的变体脚本。
- data_processing：
  - PcaptoImage-nonvpn-all.py：非 VPN 批量预处理变体。
- mnist.py：把 `png-*` 目录打包为 IDX（`train/t10k`）并压缩。
- fenxi.py：IDX 文件解析和可视化检查脚本。
- copypng.py：按比例拆分/移动 PNG（训练集/测试集整理）。
- packet-vpn、packet-nonvpn：中间 pcap 样本目录。
- png-vpn、png-nonvpn、png-10：中间 PNG 图像目录。

**历史训练线（TensorFlow 旧流程）**
- costSensitive/cnn784.py：TensorFlow v1 训练/评估主脚本（旧流程）。
- costSensitive/input_data.py：TF 时代 MNIST 数据读取工具（兼容旧代码）。
- costSensitive/model_5class_5_Mnist-10：旧 5 类 checkpoint。
- costSensitive/model_11class_5_Mnist-10：旧 11 类 checkpoint。
- out.txt：历史实验结果记录（分类指标追加日志）。

**RSG 研究子工程（扩展，不是你当前主线必须）**
- RSGmain：CVPR 2021 RSG 模块与长尾分类实验工程。
  - RSG.py：RSG 核心模块实现。
  - README.md：原论文代码使用说明。
  - Imbalanced_Classification：长尾分类实验脚本：
    - cifar_train.py：CIFAR 长尾训练入口
    - imagenet_lt_train.py、imagenet_lt_test.py、imagenet_lt_data.py：ImageNet-LT 训练/测试/数据
    - inaturalist_train.py、inaturalist_data.py：iNaturalist 训练/数据
    - places_train.py、places_test.py、places_data.py：Places-LT 训练/测试/数据
    - imbalance_cifar.py：不平衡 CIFAR 数据生成
    - losses.py：损失函数（如 LDAM/Focal）
    - utils.py：通用工具
    - models：骨干网络与 RSG 结合实现
      - resnet.py、resnet_cifar.py、resnext_cifar.py、densenet_cifar.py：模型结构
      - RSG.py：模型内 RSG 组件
      - utils.py、__init__.py：模型工具与导出

**数据/样本与辅助目录**
- dataset：数据镜像目录（当前与你主线中的 `5_Mnist-10` 有重叠用途）。
- 5_Mnist-10：仅 `.gz` 数据副本。
- 10lei：原始分类数据源目录（供预处理脚本遍历）。
- vpntest：VPN 示例流量文件集。
  - vpnchat：聊天相关 pcap 示例
  - vpnemail：邮件相关 pcap 示例
- analysis.py：统计某数据目录各类体量并画柱状图。
- test.py：临时张量/逻辑测试脚本。
- __pycache__：Python 缓存字节码。
- .idea：PyCharm 工程配置（与训练逻辑无关）。
- costSensitive/新建文件夹：空或临时目录（通常无功能含义）。
